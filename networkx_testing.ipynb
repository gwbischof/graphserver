```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
print(G.nodes)
print(G.edges)
```

```python
def plot(G):
    plt.subplot(121)
    nx.draw(G, with_labels=True, font_weight='bold')
```

```python
plot(G)
```
